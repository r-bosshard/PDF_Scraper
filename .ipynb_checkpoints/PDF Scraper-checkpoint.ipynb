```python
df=df[1]
df.columns = ["End-Market","Revenue Growth","Pricing","Quarter Forecast", "Full Year Forecast"]
df = df.set_index("End-Market")
df
```

```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook

path = r"test.xlsx"

book = load_workbook(path)
writer = pd.ExcelWriter(path, engine = 'openpyxl')
writer.book = book


df.to_excel(writer, sheet_name = 'x3')

writer.save()
writer.close()
```

```python
import pandas as pd
from openpyxl import load_workbook

path = r'test.xlsx'

writer = pd.ExcelWriter(path, engine='openpyxl')
book = load_workbook(path)
writer.book = book
writer.sheets = dict((ws.title, ws) for ws in book.worksheets)

df.to_excel(writer, sheet_name='x3', header=True, index=True,
             startcol=0,startrow=20)

writer.save()
```
